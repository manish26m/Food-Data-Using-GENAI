```python
extactfunc=[
    {
        'name': 'extractfood',
        'description': 'Get the user information from the body of the input text, add None if you get no information of any paramenter',
        'parameters': {
            'type': 'object',
            'properties': {
                'name': {
                    'type': 'string',
                    'description': 'Name of the person'
                },
                'fav_food': {
                    'type': 'string',
                    'description': 'the favorite food'
                },
                'fav_restaurant': {
                    'type': 'string',
                    'description': 'the favorite resturant'
                },
                'total_amt': {
                    'type': 'string',
                    'description': 'amount spent in example:- Rs. 5000'
                },
                'city': {
                    'type': 'string',
                    'description': 'which city is he in'
                }
                
                
            }
        }
    }
]
```

```python
suginfo=[
    {
        'name': 'extractsuggestion',
        'description': 'suggestion of food resturant',
        'parameters': {
            'type': 'object',
            'properties': {
                'suggestion': {
                    'type': 'string',
                    'description': f'suggestion of food resturant nearby'
                }
                }
                  
            }
        }
    
]
```

```python
data=[]
c=0
for i in range(len(df)):
    try:
    
        prompt=df[i]
        userinfo=client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role":"user","content":prompt}],
        functions=extactfunc
        )
        loading=userinfo.choices[0].message.function_call.arguments
        m=json.loads(loading)
    
        city=m["city"]
        food=m["fav_food"]
        p = f''' my favourite food {food} and I live in {city}, now give me a name of restraunt in the city which make best my favourite food and only give me the name of restraunt as output
        '''
        suggestioninfo=client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role":"user","content":p
            }],
            )    

            
        suggestions=suggestioninfo.choices[0].message.content
        
        m.update({"suggestion":suggestions})
        #dfm=pd.DataFrame(m)
        l=m
    except Exception as e:
        c+=1
        l=json.loads(loading)
        l.update({"fav_food":"None","suggestion":"None"})
        continue
    data.append(l)
```
